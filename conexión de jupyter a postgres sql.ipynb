```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import psycopg2
import warnings
import os
from dotenv import load_dotenv

# Ignorar todos los warnings
warnings.filterwarnings('ignore')
```

#### Crear un archivo .env
En el mismo se deben cargar todos las variables de entorno del proyecto para que luego sean levantadas por dotenv. 

```python
# Cargar las variables de entorno desde el archivo .env
load_dotenv()

# Leer las variables de entorno
db_name = os.getenv('POSTGRES_DB')
db_user = os.getenv('POSTGRES_USER')
db_password = os.getenv('POSTGRES_PASSWORD')
db_host = os.getenv('POSTGRES_HOST')
db_port = os.getenv('POSTGRES_PORT')
```

```python
# Configurar la conexión a PostgreSQL
conn = psycopg2.connect(
    dbname=db_name,
    user=db_user,
    password=db_password,
    host=db_host, 
    port=db_port       
)

# Ejecutar la consulta SQL y cargar los resultados en un DataFrame de Pandas
df = pd.read_sql_query("SELECT * FROM public.netflix", conn)

# Cerrar la conexión
conn.close()

df
```

```python
df["genre"] = df["listed_in"].str.split(", ")
df.head()
```

```python
df.info()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
df["release_year"] = df["release_year"].fillna(0)
df = df.astype({
    "release_year": "int64"
})
df.iloc[:,:7] = df.iloc[:,:7].astype("category")
# df.iloc[:,8:] = df.iloc[:,8:].astype("category")

df.info()
```

```python
df["year_added"] = df.date_added.dt.year
df.head()
```

```python
# Count de titulos por type y date_added
df2 = df.groupby(["year_added","type"])["show_id"].count().sort_values(ascending=False).reset_index()
df2
```

```python
df.info()
```

```python
df3 =  df.groupby(["country","year_added","type"])["title"].count().reset_index().sort_values(by="title",ascending=False)
df3
```

```python
from sqlalchemy import create_engine

# Crear una conexión con SQLAlchemy
engine = create_engine(f'postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')

# Guardar el DataFrame manipulado en una nueva tabla en PostgreSQL
df2.to_sql('netflix_agrupado_2', engine, index=False, if_exists='replace')

print("Datos guardados en la nueva tabla en PostgreSQL")
```
